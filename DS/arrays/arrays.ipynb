# Arrays

An array is a data structure that can hold a list of **contiguous** elements, typically of the same data type. 
- Fixed Size: The size of an array is defined at the time of its creation and cannot be changed.
- Indexed: Elements in an array can be accessed using their index, which usually starts from 0.
- dynamic arrays: Some programming languages support dynamic arrays that can grow or shrink in size as needed.


```python
# Original array
arr = [1, 2, 3]
print("initial array: ", arr)
```

```python
# appending at the end - O(1)
arr.append(4)
print("adding the value 4 to array: ", arr)
```

```python
# deleting at the end - O(1)
arr.pop()
print("deleting the last element of array: ", arr)
```

```python
#  Inserting at a specific index - O(n)
arr.insert(2, 5)
print("inserting the value 5 at index 2: ", arr)
```

```python
# Modify an element by index - O(1)
arr[3] = 5
print("Modify array at index 3: ", arr)
```

```python
# Accessing an element at a given index - O(1)
print("Accessing array at index 3: ", arr[3])
```

```python
# Checking if an element exist in the array - O(n)
# Returns the first finding
num = 9
if num in arr:
    print(f"Array contains the value {num}")
else:
    print(f"Array does not contain the value {num}")
```

```python
# Checkin length - O(1)
print("Lengt of the array: ", len(arr))
```
